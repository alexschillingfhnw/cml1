```python
# create new column with price_s as double value 
df['price_search'] = df['price_s'].str.replace('[^\d.]', '', regex=True).astype(float)
df['price_search']
```

```python
# merge price_cleaned to price_search
df['price_search'] = df['price_search'].fillna(df['price_cleaned'])
```

```python
# count how often price_search and price_cleaned are not equal
(df['price_search'] != df['price_cleaned']).sum()
```

```python
numeric_mask = df['price_search'].notna() & df['price_cleaned'].notna()

# Create a boolean mask for rows where the values in the two columns are not the same
not_equal_mask = df['price_search'] != df['price_cleaned']

# Combine the two masks using the & operator
final_mask = numeric_mask & not_equal_mask

# Use the final mask to filter the DataFrame
df[final_mask]
```

### Data Cleaning

```python
# Complete Merges
df['Floor_space_merged'] = df['Floor_space_merged'].fillna(df['Floor space:'])
df['Plot_area_merged'] = df['Plot_area_merged'].fillna(df['Land area:'])
```

```python
# remove ' m²' and ' m2' from 'Living_space_merged', 'Plot_area_merged', 'Floor_space_merged' in df
df['Living_space_merged'] = df['Living_space_merged'].str.replace(' m²', '')
df['Plot_area_merged'] = df['Plot_area_merged'].str.replace(' m²', '')
df['Floor_space_merged'] = df['Floor_space_merged'].str.replace(' m²', '')
df['Floor_space_merged'] = df['Floor_space_merged'].str.replace(' m2', '')
df['Plot_area_merged'] = df['Plot_area_merged'].str.replace(' m2', '')
# replace commas with '' in those Living_space_merged
df['Living_space_merged'] = df['Living_space_merged'].str.replace(',', '')
df['Plot_area_merged'] = df['Plot_area_merged'].str.replace(',', '')
df['Floor_space_merged'] = df['Floor_space_merged'].str.replace(',', '')

# try converting those Living_space_merged to float
df['Living_space_merged'] = df['Living_space_merged'].astype(float)   
df['Plot_area_merged'] = df['Plot_area_merged'].astype(float)
df['Floor_space_merged'] = df['Floor_space_merged'].astype(float)
```

```python
# merge Gross return
df['Gross return'] = df['Gross return'].fillna(df['Gross yield:'])
```

```python
def distplot(df, col, b):
    sns.displot(df[col], kde = True, aspect = 1.5, bins = b)
    plt.title(col)
    plt.show()

def scatterplot(df, col1, col2):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x=col1, y=col2)
    plt.title(col1 + ' vs ' + col2)
    plt.show()
```
